```python

import numpy as np 
import pandas as pd
from pandas.plotting import scatter_matrix
from matplotlib import pyplot
```

```python
project = pd.read_csv("2018RegularSeasonDetailedStats_Cleaned (1).csv")
project
```

```python
print(project.columns)
project.head()
```

```python
X = project.iloc[:,6:] 
print(X)
y = project.loc[:,'TScore']
print(y)
X_opp = project.iloc[:,6:]
y_opp = project.loc[:,'OScore']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=123)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_opp, y_opp, test_size=0.33, random_state=123)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from mlxtend.plotting import plot_decision_regions
Classifier = DecisionTreeClassifier(criterion='entropy',random_state=1,max_depth=3)
clf = Classifier.fit(X_train,y_train)
clf
clf_opp = Classifier.fit(X_train2,y_train2)
clf_opp
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import mean_squared_error
y_pred = Classifier.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
mse = mean_squared_error(y_pred,y_test)
rmse = np.sqrt(mse)
print(rmse)

y_pred_opp = Classifier.predict(X_test2)
print("Accuracy:",metrics.accuracy_score(y_test2, y_pred_opp))
print(confusion_matrix(y_test2,y_pred_opp))
print(classification_report(y_test2,y_pred_opp))
mse_opp = mean_squared_error(y_pred_opp,y_test2)
rmse_opp = np.sqrt(mse_opp)
print(rmse_opp)
```

```python
# Calculating impurity from the example above
# Maximum state of impurity, 1.0
Max_impurity = -0.5*np.log2(0.5) + (-0.5*np.log2(0.5))
print(Max_impurity)
# Entropy of child 1 = 0
# Perfect split for this child

# Entropy of child 2 = 0.918
entropy_child2 = -(2/3)*np.log2(2/3) - (1/3)*np.log2(1/3)
print(entropy_child2)

# Weighted average of entropy(children)
avg_entropy = (3/4)*(0.9184) + (1/4)*0
print(avg_entropy)


# Entropy Gain
gain_entropy = 1 - (3/4)*(0.9184) + (1/4)*0
print(gain_entropy)
```
